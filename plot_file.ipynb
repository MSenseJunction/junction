```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset created earlier
path = "synthetic_wsht_weather_migraine_prob_600days_hourly_FIXED.csv"
df = pd.read_csv(path, parse_dates=["timestamp"]).set_index("timestamp")

# Select the first 10 days (240 hours)
ten_days = df.iloc[:24*10]

# List of features to plot (each in its own figure)
features = [
    ("workload_0_10", "Workload (010)"),
    ("stress_0_10", "Stress (010)"),
    ("hrv_rmssd_ms", "HRV (RMSSD, ms)"),
    ("pressure_hpa", "Pressure (hPa)"),
    ("humidity_pct", "Humidity (%)"),
    ("temperature_c", "Temperature (C)"),
    ("migraine_prob_next_hour", "Migraine Probability (next hour)")
]

# Create one plot per feature
for col, label in features:
    plt.figure(figsize=(10, 3.2))
    plt.plot(ten_days.index, ten_days[col], marker='o', markersize=2.5, linewidth=1.0)
    plt.title(f"{label}  First 10 Days")
    plt.xlabel("Time")
    plt.ylabel(label)
    plt.grid(True, which="both", axis="both", linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.show()
```
